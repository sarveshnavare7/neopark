```python
df1=df
df1=df1.drop(columns=['B','C'])
df1
```

```python
df2=df
df2=df2.drop(columns=['A','C'])
df2
```

```python
df3=df
df3=df3.drop(columns=['A','B'])
df3
```

```python
df1.dropna(subset = ["A"], inplace=True)
df1=df1.sort_values("A")
df1
```
